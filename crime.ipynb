```python
#Method 3
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency
class ChiSquare:
    def __init__(self, dataframe):
        self.df = dataframe
        self.p = None #P-Value
        self.chi2 = None #Chi Test Statistic
        self.dof = None
        
        self.dfObserved = None
        self.dfExpected = None
    def _print_chisquare_result(self, colX, alpha):
        result = ""
        if self.p<alpha:
            result="\t\t It is important varaiable Keep it"
        else:
            result="\t\t It is not an important variable Discard it"

#         self_p1=print("P value : ")
#         alpha1=print("Alpha value : ")
        print(var,"P value =",self.p,"alpha =",alpha,result)
        
        #results=[({0},self.p,alpha,result).format(colX)]
       # print(tabulate(results, headers=["variable","pvalue", "alpha","results"]))
        
        
    def TestIndependence(self,colX,colY, alpha=0.05):
        X = self.df[colX].astype(str)
        Y = self.df[colY].astype(str)
        
        self.dfObserved = pd.crosstab(Y,X) 
        chi2, p, dof, expected = stats.chi2_contingency(self.dfObserved.values)
        self.p = p
        self.chi2 = chi2
        self.dof = dof 
        
        self.dfExpected = pd.DataFrame(expected, columns=self.dfObserved.columns, index = self.dfObserved.index)
        
        self._print_chisquare_result(colX,alpha)
df =crime
#Initialize ChiSquare Class
cT = ChiSquare(df)

#Feature Selection
testColumns = [ 'Block','IUCR','Description','Location.Description','Arrest','Domestic','FBI.Code']
for var in testColumns:
    cT.TestIndependence(colX=var,colY="Primary.Type" )
```

```python
#splitting into x and y
y = crime['Primary.Type'] 
x = crime.drop(['Primary.Type', 'Y'], axis = 1)
```

```python
#spliting data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=1)
```

# Decision Tree

```python
from sklearn import tree
# Create Decision Tree classifer object
clf = tree.DecisionTreeClassifier(criterion="entropy",max_depth=4)
# Train Decision Tree Classifer
clf = clf.fit(x_train, y_train)
#Predict the response for test dataset
y_pred = clf.predict(x_test)
train_pred = clf.predict(x_train)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn import metrics
print("DTree Accuracy on Test:", metrics.accuracy_score(y_test, y_pred))
print("DTree Accuracy on Train:", metrics.accuracy_score(y_train, train_pred))
```

# KNN

```python
# will Remove the var which we found not that important while building random Forest
x_train = x_train.drop(["Longitude","Latitude", "X.Coordinate", "Y.Coordinate","District", 
                       "Beat","Ward","Block","Community.Area"], axis = 1)
x_test = x_test.drop(["Longitude","Latitude", "X.Coordinate", "Y.Coordinate","District", 
                       "Beat","Ward","Block","Community.Area"], axis = 1)
```

```python
#feature scalling 
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(x_train)

x_train = scaler.transform(x_train)  
x_test = scaler.transform(x_test)  
```

```python
x_train.columns
```

```python
#knn Algorithm
from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=4,weights = 'distance')  
classifier.fit(x_train, y_train)  
#predicting y var on test
y_pred = classifier.predict(x_test)  
train_pred = classifier.predict(x_train)
```

```python
print('Accuracy On Test:', metrics.accuracy_score(y_test, y_pred))
print('Accuracy On Train:', metrics.accuracy_score(y_train, train_pred))
```

```python
error = []
# Calculating error for K values between 1 and 40
for i in range(1, 40):  
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error.append(np.mean(pred_i != y_test))
```
